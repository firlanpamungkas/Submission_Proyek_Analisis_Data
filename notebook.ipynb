## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana pola penggunaan sepeda berubah sepanjang waktu?

```python
plt.figure(figsize=(10, 6))
bike_day_df['dteday'] = pd.to_datetime(bike_day_df['dteday'])
plt.plot(bike_day_df['dteday'], bike_day_df['cnt'], marker='o', linestyle='-')
plt.title('Pola Penggunaan Sepeda Seiring Waktu')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Peminjaman Sepeda')
plt.grid(True)
plt.show()
```

### Pertanyaan 2: Bagaimana distribusi peminjaman sepeda berdasarkan cuaca (weather) dan hari kerja (workingday)?

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='weathersit', y='cnt', hue='workingday', data=bike_hour_df)
plt.title('Distribusi Peminjaman Sepeda Berdasarkan Cuaca dan Hari Kerja')
plt.xlabel('Cuaca')
plt.ylabel('Jumlah Peminjaman Sepeda')
plt.legend(title='Hari Kerja', loc='upper right')
plt.show()
```

## Conclusion

### - Conclution pertanyaan 1 : Bagaimana pola penggunaan sepeda berubah sepanjang waktu?


Berdasarkan grafik pola penggunaan sepeda seiring waktu, terlihat bahwa jumlah peminjaman sepeda cenderung meningkat selama beberapa bulan pertama dan kemudian menurun pada akhir tahun. Ini mungkin terkait dengan musim atau peristiwa tertentu.

### - Conclution pertanyaan 2 : Bagaimana distribusi peminjaman sepeda berdasarkan cuaca (weather) dan hari kerja (workingday)?

Distribusi peminjaman sepeda berdasarkan cuaca dan hari kerja menunjukkan bahwa pada hari kerja, jumlah peminjaman sepeda cenderung lebih tinggi pada cuaca yang lebih baik (weathersit = 1 atau 2). Namun, pada hari libur (workingday = 0), distribusi peminjaman tidak terlalu dipengaruhi oleh kondisi cuaca.

```python
bike_hour_df.info()
```

```python
# Convert 'dteday' to datetime format
bike_hour_df['dteday'] = pd.to_datetime(bike_hour_df['dteday'])

# Menentukan RFM metrics
rfm_data = bike_hour_df.groupby('registered').agg({
    'dteday': lambda x: (bike_hour_df['dteday'].max() - x.max()).days,  # Recency
    'instant': 'count',  # Frequency
    'cnt': 'sum'  # Monetary (jumlah peminjaman)
})

# Rename kolom agar lebih jelas
rfm_data.rename(columns={
    'dteday': 'Recency',
    'instant': 'Frequency',
    'cnt': 'Monetary'
}, inplace=True)

# Menampilkan beberapa baris pertama dari data RFM
print(rfm_data.head())
```

```python
# Menentukan RFM metrics
rfm_data = bike_hour_df.groupby('registered').agg({
    'dteday': lambda x: (bike_hour_df['dteday'].max() - x.max()).days,  # Recency
    'instant': 'count',  # Frequency
    'cnt': 'sum'  # Monetary (jumlah peminjaman)
})

# Rename kolom agar lebih jelas
rfm_data.rename(columns={
    'dteday': 'Recency',
    'instant': 'Frequency',
    'cnt': 'Monetary'
}, inplace=True)

# Menampilkan beberapa baris pertama dari data RFM
print(rfm_data.head())

# Menentukan kategori RFM
rfm_labels = ['1', '2', '3', '4']
rfm_data['Recency_Category'] = pd.qcut(rfm_data['Recency'], q=4, labels=rfm_labels)
rfm_data['Frequency_Category'] = pd.qcut(rfm_data['Frequency'], q=4, labels=rfm_labels)
rfm_data['Monetary_Category'] = pd.qcut(rfm_data['Monetary'], q=4, labels=rfm_labels)

# Menampilkan beberapa baris pertama dari data RFM dengan kategori
print(rfm_data.head())
```

```python
# Visualisasi RFM Analysis
plt.figure(figsize=(12, 8))

# Scatter plot Recency vs Frequency
plt.subplot(2, 2, 1)
plt.scatter(rfm_data['Recency'], rfm_data['Frequency'], alpha=0.5)
plt.title('Recency vs Frequency')
plt.xlabel('Recency (days)')
plt.ylabel('Frequency')

# Scatter plot Frequency vs Monetary
plt.subplot(2, 2, 2)
plt.scatter(rfm_data['Frequency'], rfm_data['Monetary'], alpha=0.5)
plt.title('Frequency vs Monetary')
plt.xlabel('Frequency')
plt.ylabel('Monetary')

# Scatter plot Recency vs Monetary
plt.subplot(2, 2, 3)
plt.scatter(rfm_data['Recency'], rfm_data['Monetary'], alpha=0.5)
plt.title('Recency vs Monetary')
plt.xlabel('Recency (days)')
plt.ylabel('Monetary')

# Scatter plot 3D Recency, Frequency, Monetary
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(rfm_data['Recency'], rfm_data['Frequency'], rfm_data['Monetary'], alpha=0.5)
ax.set_xlabel('Recency (days)')
ax.set_ylabel('Frequency')
ax.set_zlabel('Monetary')
ax.set_title('3D Scatter Plot - RFM Analysis')

plt.show()
```
